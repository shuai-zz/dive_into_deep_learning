# 数据操作

## 基础

- 张量(tensor)：n维数组，一个轴的tensor -> vector，两个轴的tensor -> matrix

```python
import torch
```

```python
# arange(start,end,step,...)
x=torch.arange(12)
x
```

```python
x.shape
```

```python
# numbers of element
x.numel()
```

```python
# reshape tensor
X=x.reshape(3,4)
# Automatically calculate the size of dimension
Y=x.reshape(-1,4)
X,Y
```

```python
# all zero tensor
A=torch.zeros((2,3,4))
# all one tensor
B=torch.ones((2,3,4))
A,B
```

```python
# 均值为0，标准差为1的标准高斯分布
torch.randn(3,4)
```

```python
torch.tensor([[2,1],[3,1],[4,3]])
```

## 运算符

```python
# 加 减 乘 除
x=torch.tensor([1.0,2,4,8])
y=torch.tensor([2,2,2,2])
x+y,x-y,x*y,x/y,x**y # ** 求幂操作：按位求幂
```

```python
torch.exp(x) # e的幂
```

```python
# 连接 concatenate
X=torch.arange(12,dtype=torch.float32).reshape((3,4))
Y=torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
# 0: y-coordinate  1: x-coordinate
torch.cat((X,Y),dim=0),torch.cat((X,Y),dim=1)
```

```python
# 按位比较
X==Y
```

```python
# 所有元素求和，产生一个单元素tensor
X.sum()
```
