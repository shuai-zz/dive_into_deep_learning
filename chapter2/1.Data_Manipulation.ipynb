## 运算符

```python
# 加 减 乘 除
x=torch.tensor([1.0,2,4,8])
y=torch.tensor([2,2,2,2])
x+y,x-y,x*y,x/y,x**y # ** 求幂操作：按位求幂
```

```python
torch.exp(x) # e的幂
```

```python
# 连接 concatenate
X=torch.arange(12,dtype=torch.float32).reshape((3,4))
Y=torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
# 0: y-coordinate  1: x-coordinate
torch.cat((X,Y),dim=0),torch.cat((X,Y),dim=1)
```

```python
# 按位比较
X==Y
```

```python
# 所有元素求和，产生一个单元素tensor
X.sum()
```

## 广播机制 broadcasting mechanism

通过复制实现扩展数组，对生成的数组执行按位操作

```python
a=torch.arange(3).reshape((3,1))
b=torch.arange(2).reshape((1,2))
a,b
```

```python
a+b
```

即$a+b=
\begin{bmatrix}
0&0\\

2&2\\
\end{bmatrix}
+

## 节省内存

```python
# python先计算X+Y，再分配新的内存，并使Y指向该地址
before = id(Y)
Y = Y + X
id(Y) == before
# 坏处：不必要的内存开销；其他引用可能仍指向旧内存地址
```

```python
# 原地操作
## 提前分配
Z = torch.zeros_like(Y)
print("id(Z) before:", id(Z))
Z[:] = X + Y
print("id(Z) after:", id(Z))

## 若后续不再使用X，可以直接将结果存放在X中
before = id(X)
X += Y
id(X) == before
```
